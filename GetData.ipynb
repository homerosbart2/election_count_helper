```python
import asyncio
import requests
from concurrent.futures import ThreadPoolExecutor
import json
```

```python
def GetImage(fname,site='https://resultados2019.tse.org.gt/img/201901/',path="data/"):
    try:
        r=requests.get(site+fname,timeout=(600,600))
        r.raise_for_status()       
    except requests.exceptions.RequestException as e:
        print("Failed to get "+site+fname)
        return 0
    with open(path+fname, 'wb') as f:
        f.write(r.content)
        return 1
```

```python
async def get_data_asynchronous(path="data/"):
    mesas=range(1,total_mesas+1)
    with ThreadPoolExecutor(max_workers=2) as executor:
        loop = asyncio.get_event_loop()
        tasks = [
                loop.run_in_executor(
                    executor,
                    ProcesarMesa,
                    mesa
                )
            for mesa in mesas
        ]
        res=await asyncio.gather(*tasks)
        with open(path+'results.csv', 'w') as f:
            f.write("mesa,datos,actas\n")
            for i,response in enumerate(res):
                print("Mesa {} response {}".format(mesas[i],response))
                f.write("{},{},{}\n".format(mesas[i],response[0],response[1]))
                pass
```

```python
def ObtenerDatosMesa(fname,req,site='https://resultados2019.tse.org.gt/201901/',path="data/"):
    try:
        r=requests.get(site+req,timeout=(600,600))
        r.raise_for_status()
    except requests.exceptions.RequestException as e:
        print("Failed to get "+site+req)
        return 0
    my_json = r.content.decode('utf8')
    with open(path+fname,'w') as outfile:  
        outfile.write(my_json)
    return 1
```

```python
def AbrirDatosMesa(fname):
    with open(fname) as json_file:  
        data = json.load(json_file)
        print(json.dumps(data, indent=4))
```

```python
def ProcesarMesa(mesa):
    data_name="mesa_"+'{0:06d}'.format(mesa)+'.json'
    req="api.php?mesa={}&vista=MESA".format(mesa)
    archivos_completos=0
    datos_completos=True
    print("Procesando mesa {}".format(mesa))
    for j in range(1,n_papeletas+1):
        fname='{0:06d}'.format(mesa*10+j)+'.jpg'
        #print("Getting image "+fname+" ...")
        if GetImage(fname):
            archivos_completos+=1
    if ObtenerDatosMesa(data_name,req)==0:
        datos_completos=False
    return(datos_completos,archivos_completos)
```

```python
total_mesas=21100
n_papeletas=5
site='https://resultados2019.tse.org.gt/img/201901/'
```

```python
loop = asyncio.get_event_loop()
future = asyncio.ensure_future(get_data_asynchronous())
loop.run_until_complete(future)
```
